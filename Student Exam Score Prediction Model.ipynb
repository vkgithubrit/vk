# Correction Matrix

```python
def perform_one_hot_encoding(df,column_name):
    dummies=pd.get_dummies(df[column_name],prefix=column_name)
    
    df=pd.concat([df.drop(column_name,axis=1),dummies],axis=1)
    return df
data=df.copy()
data1=df.copy()
for column in ['Gender','EthnicGroup','ParentEduc','LunchType',
               'ParentMaritalStatus','PracticeSport','WklyStudyHours']:
    data=perform_one_hot_encoding(data,column)
    
for column in ['Gender','EthnicGroup','LunchType','ParentMaritalStatus',
               'WklyStudyHours']:
    data1=perform_one_hot_encoding(data1,column)
```

```python
data1_encoded = pd.get_dummies(data1, drop_first=True)
correlation_matrix = data1_encoded.corr()
plt.figure(figsize=(30,30))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5, fmt='.2f')
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
correlation_matrix=data.corr()
for score_column in ['MathScore','ReadingScore','WritingScore','MeanScore']:
    corr=data.corr()
    target_corr=corr[score_column].drop(score_column)
    target_corr_sorted=target_corr.sort_values(ascending=False)
    plt.figure(figsize=(5,10))
    sns.heatmap(target_corr_sorted.to_frame(),cmap="coolwarm",annot=True,fmt='2f')
    plt.title(f'Corelation with {score_column}')
    plt.show()
```

# Making the Predection

```python
data=df.copy()
data.head()
```

```python
# Correct way to access columns
column_names = data.columns
```

```python
# Convert column names to a list
column_list = list(data.columns)
```

```python
column_names
```

```python
cate_val=[]
cont_val=[]

for column in data.columns:
    if data[column].nunique()<=10:
        cate_val.append(column)
    else:
        cont_val.append(column)
```

```python
cate_val
```

```python
cont_val
```

```python
for col in cate_val:
    data=pd.get_dummies(data,columns=[col],prefix=[col],drop_first=True)
```

```python
feature=data.drop(cont_val,axis=1)
target_cols=cont_val
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import r2_score, mean_squared_error,mean_absolute_error, roc_auc_score
from sklearn import svm
ss=StandardScaler()
x=ss.fit_transform(feature)
```

```python
parameters = {'C': np.logspace(-3, 2, 6), 
              'epsilon': np.logspace(-3, 2, 6), 
              'kernel': ['linear', 'poly', 'rbf', 'sigmoid']}

for target in target_cols:
    y = data[target]

    # Split the dataset into training set and test set
    x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)

    # Perform Randomized Search CV
    random_search = RandomizedSearchCV(svm.SVR(), parameters, cv=4, n_jobs=-1, verbose=2)
    random_search.fit(x_train, y_train)
    
    # Get the best estimator
    best_model = random_search.best_estimator_

    # Predict the scores
    y_pred = best_model.predict(x_test)

    # Evaluation
    r2 = r2_score(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)

    print(f"{target} - R2 Score: {r2}")
    print(f"{target} - Mean Squared Error: {mse}")
    print(f"{target} - Mean Absolute Error: {mae}")
     # Visualization: plot predicted vs actual values
    plt.figure(figsize=(12, 6))
    plt.scatter(y_test, y_pred, alpha=0.5)
    plt.xlabel("Actual Scores")
    plt.ylabel("Predicted Scores")
    plt.title(f"Actual vs Predicted Scores for {y}")
    plt.show()
```
